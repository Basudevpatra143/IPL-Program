```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Multiplayer Basketball Game")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)

# Load assets
background_image = pygame.image.load("D:/Game/51437.jpg")  # Basketball court
basket_image = pygame.image.load("C:/Users/patra/Downloads/hoop 2.png")  # Hoop
ball_image = pygame.image.load("D:/Game/ball.png")  # Basketball

# Scale assets
background_image = pygame.transform.scale(background_image, (WIDTH, HEIGHT))
basket_image = pygame.transform.scale(basket_image, (150, 100))
ball_image = pygame.transform.scale(ball_image, (60, 60))

# Basket position
basket_x, basket_y = WIDTH - 470, HEIGHT // 4

# Ball properties for Player 1 and Player 2
players = {
    1: {"ball_x": 200, "ball_y": HEIGHT - 100, "ball_speed_y": 0, "on_ground": True},
    2: {"ball_x": 600, "ball_y": HEIGHT - 100, "ball_speed_y": 0, "on_ground": True},
}

ball_width, ball_height = 60, 60
gravity = 0.5
shoot_power = -23

# Timer settings
game_time = 20  # 20 seconds per player
start_ticks = pygame.time.get_ticks()

# Scoring and game states
player1_score = 0
player2_score = 0
current_player = 1
game_active = True
winner_declared = False

# Fonts
font = pygame.font.SysFont("Arial", 48)

# Clock
clock = pygame.time.Clock()

while not winner_declared:
    screen.fill(WHITE)
    clock.tick(60)

    # Calculate elapsed time
    elapsed_seconds = (pygame.time.get_ticks() - start_ticks) // 1000
    time_left = max(game_time - elapsed_seconds, 0)

    # Draw background
    screen.blit(background_image, (0, 0))

    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    if game_active:
        # Get the current player's ball properties
        ball = players[current_player]
        ball_x, ball_y, ball_speed_y, on_ground = (
            ball["ball_x"],
            ball["ball_y"],
            ball["ball_speed_y"],
            ball["on_ground"],
        )

        # Key inputs
        keys = pygame.key.get_pressed()

        # Player 1 Controls
        if current_player == 1:
            if keys[pygame.K_LEFT] and ball_x > 0:
                ball_x -= 10
            if keys[pygame.K_RIGHT] and ball_x + ball_width < WIDTH:
                ball_x += 10
            if keys[pygame.K_SPACE] and on_ground:
                ball_speed_y = shoot_power
                on_ground = False

        # Player 2 Controls
        elif current_player == 2:
            if keys[pygame.K_LEFT] and ball_x > 0:  # A to move left
                ball_x -= 10
            if keys[pygame.K_RIGHT] and ball_x + ball_width < WIDTH:  # D to move right
                ball_x += 10
            if keys[pygame.K_SPACE] and on_ground:  # W to shoot
                ball_speed_y = shoot_power
                on_ground = False

        # Apply gravity
        if not on_ground:
            ball_speed_y += gravity
            ball_y += ball_speed_y

            # Check ground collision
            if ball_y + ball_height >= HEIGHT:
                ball_y = HEIGHT - ball_height
                ball_speed_y = 0
                on_ground = True
                game_active = True  # End turn

        # Check scoring
        if (
            basket_x < ball_x < basket_x + 150
            and basket_y < ball_y < basket_y + 100
            and not on_ground
        ):
            if current_player == 1:
                player1_score += 1
            elif current_player == 2:
                player2_score += 1
            ball_x, ball_y = 200, HEIGHT - 100  # Reset ball position
            ball_speed_y = 0
            on_ground = True
            game_active = True  # End turn

        # Update player's ball properties
        ball["ball_x"], ball["ball_y"], ball["ball_speed_y"], ball["on_ground"] = (
            ball_x,
            ball_y,
            ball_speed_y,
            on_ground,
        )

        # Draw basket
        screen.blit(basket_image, (basket_x, basket_y))

        # Draw basketball
        screen.blit(ball_image, (ball_x, ball_y))

    # Draw scores and timer
    player1_text = font.render(f"Player 1 Score: {player1_score}", True, RED)
    player2_text = font.render(f"Player 2 Score: {player2_score}", True, RED)
    timer_text = font.render(f"Time Left: {time_left}s", True, RED)
    screen.blit(player1_text, (20, 20))
    screen.blit(player2_text, (20, 80))
    screen.blit(timer_text, (WIDTH - 300, 20))

    # Handle turn switching
    if time_left == 0 or not game_active:
        if current_player == 1: 
            current_player = 2
            start_ticks = pygame.time.get_ticks()  # Reset timer
        else:
            winner_declared = True

    # Update display
    pygame.display.flip()

# Determine winner
screen.fill(WHITE)
if player1_score > player2_score:
    winner_text = font.render("Player 1 Wins!", True, GREEN)
elif player2_score > player1_score:
    winner_text = font.render("Player 2 Wins!", True, GREEN)
else:
    winner_text = font.render("It's a Tie!", True, RED)

screen.blit(winner_text, (WIDTH // 2 - 200, HEIGHT // 2))
pygame.display.flip()
pygame.time.delay(5000)

pygame.quit()
sys.exit()
```
